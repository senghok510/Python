# Histogramme

Soient $X_1,X_2,...,X_n$ des variables aléatoires i.i.d de densité $f$. La densité $f$ est inconnue et nous souhaitons l'estimer à partir de l'échantillon $X_1,X_2,...,X_n$. On se donne la densité $f$ est à support dans \[0,1]\.

On introduit un entier positif $K$ et des intervalles réguliers $I_k = \left( \frac{k-1}{K}, \frac{k}{K} \right]$, $k = 1, \ldots, K$ de longueur $\frac{1}{K}$ qui forment une partition de l’intervalle $[0, 1]$. D’abord, notons la proportion d’observations dans l’intervalle $I_k$  par :  $$ H_k := \frac{\#\{i : X_i \in I_k\}}{n}. $$

## T1. 
La loi de $H_k$, son espérance et sa variance :
    

1.  La loi de $H_k$ :

    On se donne:



    $$ H_k = \frac{1}{n} \sum_{i=1}^n \mathbf{1}_{X_i \in I_k} $$

    où $\mathbf{1}_{X_i \in I_k}$ est une variable indicatrice qui vaut 1 si $X_i \in I_k $ et 0 sinon.

    Comme $\mathbf{1}_{X_i \in I_k}$ suit loi de bernoullie:

    Ainsi, $H_k$ suit une loi Binomiale proportionnelle de paramètres $n$ et $p_k$, où $p_k $ est la probabilité qu'une observation tombe dans l'intervalle $I_k$.

    $$
    H_k \sim \frac{1}{n} \text{Binom}(n, p_k)
    $$

2. Espérance de $H_k$:

L'espérance de $H_k$ est donnée par: 
$$
\mathbb{E}[H_k] = \mathbb{E}\left[\frac{1}{n} \sum_{i=1}^n \mathbf{1}_{X_i \in I_k}\right] = \frac{1}{n} \sum_{i=1}^n \mathbb{E}[\mathbf{1}_{X_i \in I_k}]
$$
Par la linéarité de l'espérance.
et Comme:
$$
\mathbb{E}[\mathbf{1}_{X_i \in I_k}] = p_k
$$
Donc : 
$$
\mathbb{E}[H_k] = \frac{1}{n} \cdot n \cdot p_k = p_k
$$


3. Variance de $H_k$:

La variance de $H_k$ est donnée par:
$$
\text{Var}(H_k) = \text{Var}\left(\frac{1}{n} \sum_{i=1}^n \mathbf{1}_{X_i \in I_k}\right) = \frac{1}{n^2} \sum_{i=1}^n \text{Var}(\mathbf{1}_{X_i \in I_k})
$$
Par la linéarité de variance et la propriété de la variance des sommes indépendantes.

Comme $\mathbf{1}_{X_i \in I_k}$ suit une loi de Bernoulli de paramètre $p_k$, sa variance est: 

$$
\text{Var}(\mathbf{1}_{X_i \in I_k}) = p_k(1 - p_k)
$$
Donc, finalement:

$$
\text{Var}(H_k) = \frac{1}{n^2} \cdot n \cdot p_k (1 - p_k) = \frac{p_k (1 - p_k)}{n}
$$

4. Limite de $H_k$ lorsque $n$ tend vers l'infini:

D'après La loi forte de grand nombre, $H_k$ la moyenne de d'échantillon de variables indécatrices:
$$
\lim_{n \to \infty} H_k = p_k
$$
où $p_k$ est la probabilité qu'une observation tombe dans l'intervalle $I_k$ : $p_k = \int_{I_k} f(x) \, dx = \int_{\frac{k-1}{K}}^{\frac{k}{K}} f(x) \, dx $

## T2.

L'expression pour $\hat{f}(x)$ est donnée par :

$$
\hat{f}(x) = C \sum_{k=1}^K H_k \mathbf{1}_{\{x \in I_k\}}
$$

**Déterminer la constante C:**

La condition pour que $\hat{f}(x)$ soit une densité est que l'intégrale de $\hat{f}(x)$ sur tout le support soit égale à 1.

$$
\int_0^1 \hat{f}(x) \, dx = 1
$$



Pour trouver \(C\), nous devons intégrer \(\hat{f}(x)\) sur \([0,1]\) :

$$
\int_0^1 \hat{f}(x) \, dx = \int_0^1 C \sum_{k=1}^K H_k \mathbf{1}_{\{x \in I_k\}} \, dx
$$

Comme les intervalles $I_k$ sont disjoints et couvrent $[0,1]$, l'intégrale peut être décomposée en une somme d'intégrales sur chaque intervalle $I_k$ :

$$
\int_0^1 \hat{f}(x) \, dx = C \sum_{k=1}^K H_k \int_{I_k} \mathbf{1}_{\{x \in I_k\}} \, dx
$$

Cela donne :

$$
\int_0^1 \hat{f}(x) \, dx = C \sum_{k=1}^K H_k \cdot \frac{1}{K}
$$

$$
\int_0^1 \hat{f}(x) \, dx = C \cdot \frac{1}{K} \sum_{k=1}^K H_k
$$

La somme des $H_k$ sur tous les intervalles est égale à 1 car $H_k$ est la proportion d'observations dans l'intervalle $I_k$ et la somme de toutes les proportions est 1 :

$$
\sum_{k=1}^K H_k = 1
$$

Donc :

$$
\int_0^1 \hat{f}(x) \, dx = C \cdot \frac{1}{K} \cdot 1 = C \cdot \frac{1}{K}
$$

Pour que $\hat{f}(x)$ soit une densité, cette intégrale doit être égale à 1 :

$$
C \cdot \frac{1}{K} = 1
$$

D'où :

$$
C = K
$$

## T3.


**Montrons que:**    $\text{MSE}(x_0) = (\mathbb{E}[\hat{f}(x_0)] - f(x_0))^2 + \text{Var}(\hat{f}(x_0)).$




Par définition, l'erreur quadratique moyenne (MSE) au point $x_0$ est :

$$
\text{MSE}(x_0) = \mathbb{E}[(\hat{f}(x_0) - f(x_0))^2]
$$

Puisque :

$$
\mathbb{E}[(\hat{f}(x_0) - f(x_0))^2] = \mathbb{E}[(\hat{f}(x_0) - \mathbb{E}[\hat{f}(x_0)])^2] + (\mathbb{E}[\hat{f}(x_0)] - f(x_0))^2 + 2\mathbb{E}[(\hat{f}(x_0) - \mathbb{E}[\hat{f}(x_0)])(\mathbb{E}[\hat{f}(x_0)] - f(x_0))]
$$

La covariance entre $(\hat{f}(x_0) - \mathbb{E}[\hat{f}(x_0)])$ et $(\mathbb{E}[\hat{f}(x_0)] - f(x_0))$ est nulle car $(\mathbb{E}[\hat{f}(x_0)] - f(x_0))$ est une constante. 

Cela simplifie à :

$$
\text{MSE}(x_0) = \mathbb{E}[(\hat{f}(x_0) - \mathbb{E}[\hat{f}(x_0)])^2] + (\mathbb{E}[\hat{f}(x_0)] - f(x_0))^2
$$

La première partie est la variance de $\hat{f}(x_0)$ et la deuxième partie est le carré du biais de $\hat{f}(x_0)$ :

Donc, 
$$
\text{MSE}(x_0) = \text{Var}(\hat{f}(x_0)) + (\mathbb{E}[\hat{f}(x_0)] - f(x_0))^2
$$


## T4.

**Montrons que le risque quadratique est donné par:**
$$
\text{MSE}(x_0) = (K p_k - f(x_0))^2 + \frac{1}{n} K^2 p_k (1 - p_k),
$$

où $ p_k := \mathbb{P}(X_i \in I_k) = \int_{I_k} f(x) \, dx $.
Pour $ x_0 \in I_k $, c'est-à-dire $\frac{k-1}{K} < x_0 \leq \frac{k}{K}$, nous avons :
$$
\hat{f}(x_0) = K H_k
$$


Ainsi :

$$
\mathbb{E}[\hat{f}(x_0)] = \mathbb{E}[K H_k] = K \mathbb{E}[H_k] = K p_k
$$

et

$$
\text{Var}(\hat{f}(x_0)) = \text{Var}(K H_k) = K^2 \text{Var}(H_k) = K^2 \frac{p_k (1 - p_k)}{n}
$$

En substituant ces résultats dans l'expression de la MSE :

$$
\text{MSE}(x_0) = \text{Var}(\hat{f}(x_0)) + (\mathbb{E}[\hat{f}(x_0)] - f(x_0))^2
$$

Nous obtenons :

$$
\text{MSE}(x_0) = K^2 \frac{p_k (1 - p_k)}{n} + (K p_k - f(x_0))^2
$$

Cela montre que le risque quadratique est donné par :

$$
\text{MSE}(x_0) = (K p_k - f(x_0))^2 + \frac{1}{n} K^2 p_k (1 - p_k).
$$

**Calcul de l'Espérance de $\hat{f}(x_0)$**

Nous avons déjà établi que :

$$
\hat{f}(x_0) = K H_k
$$

L'espérance de $\hat{f}(x_0)$ est donc :

$$
\mathbb{E}[\hat{f}(x_0)] = \mathbb{E}[K H_k] = K \mathbb{E}[H_k]
$$

$$
\mathbb{E}[\hat{f}(x_0)] = K p_k
$$

où $ p_k = \int_{I_k} f(x) \, dx $.

Donc, $\hat{f}(x_0)$ n'est pas un estimateur sans biais de $f(x_0)$.      


## T5.


Pour $ n $ tendant vers l'infini :

- **Biais**: Le biais est donné par $ K p_k - f(x_0) $. Ce terme ne dépend pas de $ n $, donc le biais reste constant lorsque $ n $ tend vers l'infini.
  
- **Variance**: La variance est donnée par $ \frac{1}{n} K^2 p_k (1 - p_k) $. Comme $ n $ tend vers l'infini, $ \frac{1}{n} $ tend vers 0, donc la variance tend vers 0.

- **Risque quadratique (MSE)**: La MSE est la somme du carré du biais et de la variance. Comme la variance tend vers 0, la MSE tend vers le carré du biais. Donc, pour $ n $ tendant vers l'infini :

$$
\text{MSE}(x_0) \to (K p_k - f(x_0))^2
$$

## T6.

Supposons $ f $ continue. Étudier la limite du biais, de la variance et du risque MSE $\text{MSE}(x_0)$ quand $n$ et $ K $ tendent vers l’infini simultanément. 

Pour que $\hat{f}(x_0)$ converge en probabilité vers $ f(x_0) $, il faut que le biais et la variance tendent vers 0.

- **Biais**: Le biais est $ K p_k - f(x_0) $. Si $ K $ tend vers l'infini, les intervalles $I_k$ deviennent de plus en plus petits. Si $ f $ est continue, alors pour des $ K $ suffisamment grands, $ p_k \approx \frac{f(x_0)}{K} $. Ainsi,

$$
K p_k \approx K \cdot \frac{f(x_0)}{K} = f(x_0)
$$

Donc, le biais $ K p_k - f(x_0) $ tend vers 0.

- **Variance**: La variance est $ \frac{1}{n} K^2 p_k (1 - p_k) $. Si $K $ et $n $ tendent vers l'infini de telle sorte que $ \frac{K^2}{n} $ reste borné, alors la variance tend vers 0.

- **MSE**: Comme le biais et la variance tendent vers 0, la MSE tend vers 0.

Ainsi, si $ K $ et $ n $ tendent vers l'infini de manière appropriée, $\hat{f}(x_0)$ converge en probabilité vers $ f(x_0) $.


## Analyse Numérique

## S1.
Dans cette partie, on va écrire une fonction pour calculer l'histogramme $\hat{f_n}$

```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as sps
from scipy.special import gamma
from scipy.stats import beta
```

```python
def histogramme_estimateur(donne, K):
   
    n = len(donne)  
    histogramme = np.zeros(K)  
    intervalle_taille = 1 / K  

    for i in range(K):
        gauche_borne = i * intervalle_taille 
        droite_borne = (i + 1) * intervalle_taille 
        nombre = np.sum((donne > gauche_borne) & (donne <= droite_borne))
        histogramme[i] = nombre / (n * intervalle_taille)
    
    return histogramme
        


```

## S2.



```python

def beta_densite(donne, a, b):
    beta_fonction = (gamma(a) * gamma(b)) / gamma(a + b)
    densité = (1 / beta_fonction) * (donne ** (a - 1)) * ((1 - donne) ** (b - 1))
    return densité



def plot_histogramme_et_densite(donne, K, a, b):
    histogramme_valeur = histogramme_estimateur(donne, K)
    x_hist = np.linspace(0, 1, K, endpoint=False) + 1/(2*K)
    donné_trié = np.sort(donne)
    beta_densite_valeurs = beta_densite(donné_trié, a, b)

    plt.bar(x_hist, histogramme_valeur, width=1/K, edgecolor='k', alpha=0.7, label='Histogramme Estimateur')
    plt.plot(donné_trié, beta_densite_valeurs, color='r', label='Densité de Beta')
    plt.xlabel('Intervalle')
    plt.ylabel('Densité')
    plt.title('Histogramme Estimateur et Densité de Belta')
    plt.legend()
    plt.show()

# Test de la fonction
a, b = 3, 2  
n = 10000 
K = 20  


donné = np.random.beta(a, b, size=n)


plot_histogramme_et_densite(donné, K, a, b)
```

## S3.

```python
def plot_histogramme_et_densite(donne, K, a, b, n):
   
    histogramme_valeur = histogramme_estimateur(donne, K)
    
    x_hist = np.linspace(0, 1, K, endpoint=False) + 1/(2*K)
    donné_trié = np.sort(donne)
    beta_densite_valeurs = beta_densite(donné_trié, a, b)
    
    plt.bar(x_hist, histogramme_valeur, width=1/K, edgecolor='k', alpha=0.7, label='Histogramme Estimateur')
    plt.plot(donné_trié, beta_densite_valeurs, color='r', label='Densité de Belta')
    plt.xlabel('Intervalle')
    plt.ylabel('Densité')
    plt.title(f'Histogramme et densité de Belta (n={n}, K={K})')
    plt.legend()
    plt.show()


a, b = 3, 2  
K = 5  


taille_échantillon = [10, 100, 1000, 10000]

for n in taille_échantillon:
    donné = np.random.beta(a, b, size=n)
    plot_histogramme_et_densite(donné, K, a, b, n)


K = 10  
a, b = 2, 5 

for n in taille_échantillon:
    donné = np.random.beta(a, b, size=n)
    plot_histogramme_et_densite(donné, K, a, b, n)
```

## Observation:

2. Quand $K$ augmente, la largeur des bins est plus petite, ce qui peut rendre l'approximation plus exacte.

# S5.

```python
def estimate_densite_a_x0(hist, K, x0):
    intervalle_taille = 1 / K
    intervalle_index = int(x0 // intervalle_taille)
    return hist[intervalle_index] if 0 <= intervalle_index < K else 0

def evaluate_mse(n, x0, K, a, b):
    nombre_simulations = 1000  
    estimates = []
    for _ in range(nombre_simulations):
        donné = np.random.beta(a, b, n)
        hist = histogramme_estimateur(donné, K)
        estimate = estimate_densite_a_x0(hist, K, x0)
        estimates.append(estimate)
    
    estimates = np.array(estimates)
    densité_réel = beta.pdf(x0, a, b)
    
    bias = np.mean(estimates) - densité_réel
    variance = np.var(estimates)
    mse = bias**2 + variance
    
    return bias, variance, mse
```

## S6.

```python

a, b = 3, 2
x0 = 0.5
K = 5
taille_échantillon = [10, 30, 100, 300, 1000, 3000, 10000]


resultat = []
for n in taille_échantillon:
    bias, variance, mse = evaluate_mse(n, x0, K, a, b)
    resultat.append((n, bias, variance, mse))
    print(f"n={n} -> Bias: {bias:.5f}, Variance: {variance:.5f}, MSE: {mse:.5f}")


n_valeurs = [r[0] for r in resultat]
bias_valeurs = [r[1] for r in resultat]
variance_valeurs = [r[2] for r in resultat]
mse_valeurs= [r[3] for r in resultat]

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(n_valeurs, bias_valeurs, marker='o')
plt.xscale('log')
plt.xlabel('Taille échantillon (n)')
plt.ylabel('Bias')
plt.title('Bias vs Taille échantillon')

plt.subplot(1, 3, 2)
plt.plot(n_valeurs, variance_valeurs, marker='o')
plt.xscale('log')
plt.xlabel('Taille échantillon (n)')
plt.ylabel('Variance')
plt.title('Variance vs Taille échantillon')

plt.subplot(1, 3, 3)
plt.plot(n_valeurs, mse_valeurs, marker='o')
plt.xscale('log')
plt.xlabel('Taille échantillon (n)')
plt.ylabel('MSE')
plt.title('MSE vs Taille échantillon')

plt.tight_layout()
plt.show()
```

## Observation:
Selon les graphiques, lorsque la taille d'échantillon $n$ augmente, le MSE tend à baisser jusqu'à zéro.

2. Le schéma suggère que le modèle est plus précis (MSE plus bas) dans la plage intermédiaire de $x_0$ et moins précis aux limites.

# S8.

```python


n = 100
a, b = 2, 2
x0_valeurs = [0.5, 0.1, 0.95]
K_valeurs = range(1, 20)  


optimal_K = {}

for x0 in x0_valeurs:
    print(f"Evaluation pour x0 = {x0}")
    min_mse = float('inf')
    best_K = None
    resultat = []
    for K in K_valeurs:
        bias, variance, mse = evaluate_mse(n, x0, K, a, b)
        resultat.append((K, bias, variance, mse))
        print(f"K={K} -> Bias: {bias:.5f}, Variance: {variance:.5f}, MSE: {mse:.5f}")
        if mse < min_mse:
            min_mse = mse
            best_K = K
    optimal_K[x0] = best_K
    print(f"Optimal K pour x0={x0} est {best_K} avec MSE={min_mse:.5f}\n")

    K_vals = [r[0] for r in resultat]
    bias_vals = [r[1] for r in resultat]
    variance_vals = [r[2] for r in resultat]
    mse_vals = [r[3] for r in resultat]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 3, 1)
    plt.plot(K_vals, bias_vals, marker='o')
    plt.xlabel('Nombre intervalles (K)')
    plt.ylabel('Bias')
    plt.title(f'Bias vs K pour x0={x0}')

    plt.subplot(1, 3, 2)
    plt.plot(K_vals, variance_vals, marker='o')
    plt.xlabel('Nombre intervalles (K)')
    plt.ylabel('Variance')
    plt.title(f'Variance vs K pour x0={x0}')

    plt.subplot(1, 3, 3)
    plt.plot(K_vals, mse_vals, marker='o')
    plt.xlabel('Nombre intervallles (K)')
    plt.ylabel('MSE')
    plt.title(f'MSE vs K pour x0={x0}')

    plt.tight_layout()
    plt.show()

print("Optimal K valeurs pour les valeurs x0 différentes:")
for x0, K in optimal_K.items():
    print(f"x0={x0}: K_opt={K}")
```

## Observation

La valeur de K qui minimise le MSE dépends d'un point $x_0$ choisit. 

# Choix optimal d’un K global

## T7.

Pour démontrer l'égalité 
$
\text{MISE}(\hat{f}_K) = \int_0^1 f^2(x) \, dx + J(K) \
$
, où 
$
J(K) = \frac{K}{n} - \frac{n+1}{n} \sum_{k=1}^K p_{k,K}^2 \
$
 et 
$
p_{k,K} = \int_{I_{k,K}} f(x) \, dx \
$
, procédons étapes par étapes.

D'abord, on rappelle l'expression de l'erreur quadratique moyenne en un point 
$
x_0 \
$ donnée par :
$$
\text{MSE}(x_0) = (K p_k - f(x_0))^2 + \frac{1}{n} K^2 p_k (1 - p_k), \
$$
où 
$ 
x_0 \in I_k \
$
 et 
$
p_k \
$
 est la proportion de la densité 
$
f \
$
 sur l'intervalle 
$ 
I_k \
$
.

Pour obtenir la MISE, il nous faut intégrer la MSE sur tout l'intervalle 
$
[0, 1] \
$
 :
$$
\text{MISE}(\hat{f}_K) = \int_0^1 \text{MSE}(x) \, dx. \
$$

Écrivons l'expression de la MSE en termes de 
$
p_k \
$
 sur l'intervalle 
$
I_k \
$
 :
$$
\text{MSE}(x) = (K p_k - f(x))^2 + \frac{1}{n} K^2 p_k (1 - p_k), \
$$
et intégrons cette expression sur 
$
I_k \
$
 :
$$
\int_{I_k} \text{MSE}(x) \, dx = \int_{I_k} (K p_k - f(x))^2 \, dx + \int_{I_k} \frac{1}{n} K^2 p_k (1 - p_k) \, dx. \
$$

Pour le premier terme:
$$
\int_{I_k} (K p_k - f(x))^2 \, dx = \int_{I_k} K^2 p_k^2 - 2K p_k f(x) + f(x)^2 \, dx. \
$$
Étant donné que 
$
\int_{I_k} f(x) \, dx = p_k \
$
, on obtient:
$$
\int_{I_k} -2K p_k f(x) \, dx = -2K p_k \int_{I_k} f(x) \, dx = -2K p_k^2. \
$$
Et donc:
$$
\int_{I_k} (K p_k - f(x))^2 \, dx = K^2 p_k^2 - 2K p_k^2 + \int_{I_k} f(x)^2 \, dx = K^2 p_k^2 - 2K p_k^2 + \int_{I_k} f(x)^2 \, dx. \
$$
Pour le deuxième terme:
$$
\int_{I_k} \frac{1}{n} K^2 p_k (1 - p_k) \, dx = \frac{1}{n} K^2 p_k (1 - p_k) \int_{I_k} 1 \, dx = \frac{1}{n} K^2 p_k (1 - p_k) \frac{1}{K} = \frac{K}{n} p_k (1 - p_k). \
$$

En sommant sur tous les 
$
K \
$
 intervalles:
$$
\text{MISE}(\hat{f}_K) = \sum_{k=1}^K \left( \int_{I_k} (K^2 p_k^2 - 2K p_k^2 + f(x)^2) \, dx + \frac{K}{n} p_k (1 - p_k) \right). \
$$
Or:
$$
\sum_{k=1}^K \int_{I_k} f(x)^2 \, dx = \int_0^1 f(x)^2 \, dx, \
$$
et:
$$
\sum_{k=1}^K \left( K^2 p_k^2 - 2K p_k^2 \right) = K \sum_{k=1}^K p_k^2 - 2K \sum_{k=1}^K p_k^2 = -K \sum_{k=1}^K p_k^2. \
$$
Ainsi:
$$
\text{MISE}(\hat{f}_K) = \int_0^1 f(x)^2 \, dx + \frac{K}{n} \sum_{k=1}^K p_k (1 - p_k) - K \sum_{k=1}^K p_k^2. \
$$

On peut réécrire le terme de correction:
$$
\frac{K}{n} \sum_{k=1}^K p_k (1 - p_k) = \frac{K}{n} \sum_{k=1}^K (p_k - p_k^2) = \frac{K}{n} \left( 1 - \sum_{k=1}^K p_k^2 \right), \
$$
car 
$
\sum_{k=1}^K p_k = 1 \
$
.

Finalement, on obtient:
$$
\text{MISE}(\hat{f}_K) = \int_0^1 f(x)^2 \, dx + \frac{K}{n} - \frac{K}{n} \sum_{k=1}^K p_k^2 - K \sum_{k=1}^K p_k^2. \
$$
En combinant les termes:
$$
\text{MISE}(\hat{f}_K) = \int_0^1 f(x)^2 \, dx + \frac{K}{n} - \left( \frac{K}{n} + K \right) \sum_{k=1}^K p_k^2. \
$$
Soit:
$$
\text{MISE}(\hat{f}_K) = \int_0^1 f(x)^2 \, dx + \frac{K}{n} - \frac{n+1}{n} K \sum_{k=1}^K p_k^2. \
$$
D'où la formule finale:
$$
\text{MISE}(\hat{f}_K) = \int_0^1 f^2(x) \, dx + J(K), \
$$
avec:
$$
J(K) = \frac{K}{n} - \frac{n+1}{n} K \sum_{k=1}^K p_{k,K}^2. \
$$
Cela complète la démonstration.

```python
def calcule_pkK(a, b, K):
    pkK = []
    beta_dist = sps.beta(a, b)
    intervals = np.linspace(0, 1, K + 1)
    for i in range(K):
        pkK.append(beta_dist.cdf(intervals[i + 1]) - beta_dist.cdf(intervals[i]))
    return np.array(pkK)

def calcule_JK(a, b, n, Kmax):
    J = []
    for K in range(1, Kmax + 1):
        pkK = calcule_pkK(a, b, K)
        JK = (K / n) - ((n + 1) / n) * K * np.sum(pkK ** 2)
        J.append(JK)
    return np.array(J)

def trouver_Kopt(a, b, n, Kmax):
    J = calcule_JK(a, b, n, Kmax)
    Kopt = np.argmin(J) + 1
    return Kopt, J

def plot_JK(J):
    plt.plot(range(1, len(J) + 1), J, marker='o')
    plt.xlabel('K')
    plt.ylabel('J(K)')
    plt.title('J(K) en fonction de K')
    plt.grid(True)
    plt.show()

# Paramètres
a, b = 2, 2
n = 100
Kmax = 50

# Calcule Kopt
Kopt, J = trouver_Kopt(a, b, n, Kmax)
print(f"La valeur optimal de K est : {Kopt}")

# Plot J(K)
plot_JK(J)

# Variation des paramètres
parameters = [(2, 5, 100), (5, 2, 100), (2, 2, 200), (3, 3, 100)]
for a, b, n in parameters:
    Kopt, J = trouver_Kopt(a, b, n, Kmax)
    print(f"Kopt pour Beta({a}, {b}) et n = {n}: {Kopt}")
    plot_JK(J)
```

## S11.

```python
def calcule_Hk(data, K):
    n = len(data)
    Hk, _ = np.histogram(data, bins=K, range=(0, 1))
    return Hk / n

def calcule_J_empirique(data, n, Kmax):
    J = []
    for K in range(1, Kmax + 1):
        Hk = calcule_Hk(data, K)
        JK = (K / n) - ((n + 1) / n) * K * np.sum(Hk**2 - Hk*(1-Hk)/n)
        J.append(JK)
    return np.array(J)

def trouver_Kopt_empirique(data, Kmax):
    n = len(data)
    J = calcule_J_empirique(data, n, Kmax)
    Kopt = np.argmin(J) + 1
    return Kopt, J

def simulation_de_monte_carlo(a, b, n, Kmax, num_simulations):
    Kopts = []
    J_values = np.zeros((num_simulations, Kmax))
    
    for i in range(num_simulations):
        data = np.random.beta(a, b, n)
        Kopt, J = trouver_Kopt_empirique(data, Kmax)
        Kopts.append(Kopt)
        J_values[i, :] = J
    
    Kopts = np.array(Kopts)
    J_means = np.mean(J_values, axis=0)
    J_stds = np.std(J_values, axis=0)
    
    return Kopts, J_means, J_stds

def calcule_pkK(a, b, K):
    pkK = []
    beta_dist = sps.beta(a, b)
    intervals = np.linspace(0, 1, K + 1)
    for i in range(K):
        pkK.append(beta_dist.cdf(intervals[i + 1]) - beta_dist.cdf(intervals[i]))
    return np.array(pkK)

def calcule_JK(a, b, n, Kmax):
    J = []
    for K in range(1, Kmax + 1):
        pkK = calcule_pkK(a, b, K)
        JK = (K / n) - ((n + 1) / n) * K * np.sum(pkK ** 2)
        J.append(JK)
    return np.array(J)

def trouver_Kopt(a, b, n, Kmax):
    J = calcule_JK(a, b, n, Kmax)
    Kopt = np.argmin(J) + 1
    return Kopt, J

# Paramètres pour la simulation Monte Carlo
a, b = 2, 2
n = 100
Kmax = 50
num_simulations = 1000

# Calcule Kopt théorique
Kopt, J = trouver_Kopt(a, b, n, Kmax)

def plot_résulat_monte_carlo(J_means, J_stds, Kmax):
    """
    On plot la moyenne et l'intervale de confiance à 95% pour J(K) ainsi que la courbe théorique.
    """
    K_range = np.arange(1, Kmax + 1)
    plt.figure(figsize=(10, 5))
    plt.plot(K_range, J_means, marker='o', label='Moyenne J(K)')
    plt.plot(range(1, len(J) + 1), J, marker='_', label='J(K) théorique')
    plt.fill_between(K_range, J_means - 1.96 * J_stds, J_means + 1.96 * J_stds, color='b', alpha=0.2, label='IC 95%')
    plt.xlabel('K')
    plt.ylabel('J(K)')
    plt.title('Simulation Monte Carlo de J(K)')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_distribution_Kopt(Kopts, Kmax):
    """
    On plot la distribution de Kopt depuis la simulation de Monte Carlo.
    """
    plt.figure(figsize=(10, 5))
    plt.hist(Kopts, bins=range(1, Kmax + 2), align='left', rwidth=0.8)
    plt.xlabel('Kopt')
    plt.ylabel('Fréquence')
    plt.title('Distribution de Kopt à partir de la Simulation Monte Carlo')
    plt.grid(True)
    plt.show()

# simulation Monte Carlo
Kopts, J_means, J_stds = simulation_de_monte_carlo(a, b, n, Kmax, num_simulations)

# Plot les résultats
plot_résulat_monte_carlo(J_means, J_stds, Kmax)
plot_distribution_Kopt(Kopts, Kmax)

# Résumé des résutats de Kopt
unique, counts = np.unique(Kopts, return_counts=True)
Kopt_summary = dict(zip(unique, counts))
print("distribution de Kopt à partir de la simulation :")
for K, count in Kopt_summary.items():
    print(f"K = {K}: {count} fois ({count / num_simulations:.2%})")
print(f"La valeur optimal de K théorique est : {Kopt}")
```

On observe bien que la courbe de la simulation de monte carlo de J(k) moyen tends vers J(k) théorique.
